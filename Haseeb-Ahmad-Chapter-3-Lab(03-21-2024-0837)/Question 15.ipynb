```python
#giving predictors  and responce 
design=MS(['zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'lstat', 'medv'])
X =design.fit_transform(Boston)
y = Boston['crim']
#model fitting
model = sm.OLS(y, X)
results = model.fit()
results.summary()
summarize(results)
```
